```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import LabelEncoder
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
raw_data = pd.read_csv("/content/gdrive/MyDrive/EAFP_files/data.csv")
```

```python
raw_data
```

```python
#looking to see if any null values
null_entries = raw_data.isna().sum().sort_values(ascending=False) / len(raw_data)
null_entries
```

```python
#checking if any duplicate employees (EmployeeNumber is a unique identifier for each employee)
raw_data.duplicated(subset="EmployeeNumber").sum()
```

```python
raw_data.dtypes
```

```python
#assigning all categorical variables numerical values instead
label_encoder = LabelEncoder()
categorical_cols = ["Attrition", 'BusinessTravel', 'Department', 
                       'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 
                       'OverTime']
raw_data[categorical_cols] = raw_data[categorical_cols].apply(label_encoder.fit_transform)
raw_data[categorical_cols]
```

```python
raw_data['rate_avg'] = raw_data[['HourlyRate', 'DailyRate', 'MonthlyRate']].mean(axis=1)
```

```python
clean_data = raw_data.copy()
clean_data.columns
```

```python
#renaming column names
clean_data.rename(columns=lambda x: re.sub(r'(?<!^)(?=[A-Z])', '_', x), inplace=True)
clean_data.columns = clean_data.columns.str.lower()
clean_data.columns
```

```python
clean_data.to_csv('clean_data.csv')
```
